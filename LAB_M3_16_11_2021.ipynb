```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1=pd.read_csv('/content/drive/MyDrive/Datasets/Melbourne_housing_FULL.csv')
```

```python
df1
```

```python
col_to_fill=['Suburb', 	'Address',	'Rooms',	'Type', 'Price',	'Method',	'SellerG',	'Date',	'Distance',	'Postcode', 	'Bedroom2',	'Bathroom',	'Car',	'Landsize', 'BuildingArea',	'YearBuilt',	'CouncilArea',	'Lattitude',	'Longtitude',	'Regionname',	'Propertycount']
```

```python
df1.isna().sum()
```

```python
df1[col_to_fill]=df1[col_to_fill].fillna(0);
df1.isna().sum()
```

```python
df1 = df1.drop(['Address','Date','Postcode','Longtitude','Lattitude','YearBuilt'],axis=1)
```

```python
df2 = pd.get_dummies(df1)
df2
```

```python
y=df2['Price']
X=df2.drop(['Price'],axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=40)
```

```python
from sklearn.linear_model import LinearRegression
LR=LinearRegression().fit(X_train,y_train)
print(LR.score(X_train,y_train))
print(LR.score(X_test,y_test))
```

```python
from sklearn.linear_model import Lasso
lasso =Lasso(alpha=0.1,tol=20)
lasso.fit(X_train,y_train)
print(lasso.score(X_train,y_train))
print(lasso.score(X_test,y_test))
```

```python
from sklearn.linear_model import Ridge
Rg = Ridge(alpha=0.1,tol=20)
Rg.fit(X_train,y_train)
print(Rg.score(X_train,y_train))
print(Rg.score(X_test,y_test))
```
